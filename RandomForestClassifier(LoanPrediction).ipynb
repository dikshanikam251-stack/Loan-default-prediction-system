```python
#4. Feature Creation
#convert categorical (text) values into numeric values
df.replace({
    "Loan_Status":{"N":0, "Y":1},
    "Gender": {"Male":0, "Female":1},
    "Education": {"Not Graduate":0, "Graduate":1},
    "Married":{"No":0, "Yes":1},
    "Self_Employed":{"No":0, "Yes":1}
}, inplace=True)
```

```python
#5. Train Test split the data
def train_test_split_and_features(df):
    y=df["Loan_Status"]              #Loan_Status is target variable
    x=df.drop(["Loan_Status", "Loan_ID"], axis=1)
    x=pd.get_dummies(data=x, columns=["Property_Area", "Dependents"])           #Converts Property_Area and Dependents into dummy columns
    
    #20% data for testing,  80% for training,random_state=0 â†’ ensures same split every time
    x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=0)
    
    print(x.head(5))
    print(x.columns)         #Print Column Names
    features=list(x.columns)      #Store Feature Names
    return x_train, x_test, y_train, y_test, features
    
```

```python
#calling the function
x_train, x_test, y_train, y_test, features=train_test_split_and_features(df)
```

```python
#6. Fit & evaluate model
def fit_evaluate_model(x_train, x_test, y_train, y_test):
    random_forest=RandomForestClassifier(random_state=0,               #Fixes randomness, Ensures same results every time you run the code.
                                         max_depth=5,                  #Maximum depth of each tree
                                         min_samples_split=0.2,       #Minimum number of samples required to split a node
                                         max_features=0.8,              #Each tree will use 80% of the features randomly.    
                                         max_samples=0.8                #Each tree is trained on 80% of the training data.   
                                        )
    ##Trains the Random Forest model on training data
    model=random_forest.fit(x_train, y_train)                            
    random_forest_predict=random_forest.predict(x_test)
    random_forest_conf_matrix=confusion_matrix(y_test, random_forest_predict)
    random_forest_acc_score=accuracy_score(y_test, random_forest_predict)
    print("confusion matrix")
    print(random_forest_conf_matrix)
    print("\n")
    print("Accuracy of Random Forest:", random_forest_acc_score*100, "\n")
    print(classification_report(y_test, random_forest_predict))
    return model
```

```python
model = fit_evaluate_model(x_train, x_test, y_train, y_test)
```

```python
#TN = 8 â†’ Correctly predicted Non-Default (0)
#FP = 7 â†’ Predicted Default but actually Non-Default
#FN = 1 â†’ Predicted Non-Default but actually Default
#TP = 46 â†’ Correctly predicted Default (1)
```

```python
# Risk score usually comes from the probability of default predicted by a machine learning model.
# This is a numerical value that indicates how risky a loan applicant is.
#Higher score â†’ Higher chance of default â†’ Higher risk
#Lower score â†’ Lower chance of default â†’ Lower risk
```

```python
import pandas as pd

#predict_proba(x_test)â†’ gives probability for each class 0 & 1
#binary classification probability of column 0->no default and column 1->default
#[:, 1] means: select all rows, Take only column 1 So will get probability of default for each loan.
#y_prob become predicted probability of 1D array
y_prob = model.predict_proba(x_test)[:, 1]       

#This creates new table
#x_test.index gives the index values of our test dataset.
results = pd.DataFrame({
    'Loan_ID': x_test.index,
    'Default_Probability': y_prob
})

def categorize_risk(prob):         # Much easier for decision-making:
    if prob > 0.7:          
        return "High"              #High â†’ Reject loan
    elif prob >= 0.4:
        return "Medium"            #Medium â†’ Review manually
    else:
        return "Low"               #Low â†’ Approve loan

        
#This adds a new column with High / Medium / Low labels.
results['Risk_Level'] = results['Default_Probability'].apply(categorize_risk)

def recovery_action(risk):
    if risk == "High":
        return "Immediate intervention"
    elif risk == "Medium":
        return "Monitoring"
    else:
        return "Standard process"
        
#This adds a new column with risk levels
results['Recovery_Action'] = results['Risk_Level'].apply(recovery_action)

print(results.head())
```
